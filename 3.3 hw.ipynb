```python
def assigning_categories(x):
    if x >= 8:
        return "High"
    elif x >= 6 and x < 8:
        return "Medium"
    else:
        return "Low"
    
```

```python
df["Category"] = df.Rating.apply(assigning_categories)
df
```

```python
procent_category = df.Category.value_counts()
length = len(df.Category)
result = procent_category / length * 100
result
```
